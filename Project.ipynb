```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the insurance sataset
df=pd.read_csv('insurance.csv')
df.head()
```

```python
# Shape of the data
print('Shape of the data: ',df.shape,'\n')

# Data types of column
print('Data types of column:')
print(df.dtypes)
```

```python
# Missing Values
print(df.isnull().sum())
```

```python
# The relationship between the feature and target column 
# using a count plot of categorical columns
cat_columns = ['sex', 'smoker', 'region', 'children']
for column in cat_columns:
    sns.countplot(data=df, x=column)
    plt.title(f'Count Plot of {column}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Scatter plot of numerical columns
num_columns = ['age', 'bmi', 'children']
for column in num_columns:
    sns.scatterplot(data=df, x=column, y='charges', hue='smoker')
    plt.title(f'Scatter Plot of {column} vs Charges')
    plt.show()
```

```python
# Data visualization using plots of feature vs feature
sns.pairplot(df, hue='smoker')
plt.show()
```

```python
sns.lmplot(data=df, x='age', y='charges', hue='smoker', palette='Set1', height=6)
plt.title("Charges vs Age by Smoking Status")
plt.show()
```
